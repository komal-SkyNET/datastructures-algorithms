```
#recursive DP algo
arr = [1,2,3,1,6,10] 
arr = [1,2,3,4,5,6]
# arr = [100000, 100000, 1, 3, 4, 5, 100000, 3, 5, 100000]
k = 2
mem = {}
def solution(pos, c):
  if (pos, c) not in mem: 
    if pos == len(arr):
      return 0
    if c == k:
      return solution(pos+1, 0)
    mem[pos, c] = max(arr[pos] + solution(pos+1, c+1), solution(pos+1, 0))
  return mem[pos, c]
print(solution(0, 0))
mem
```

```
#observe in a matrix
import numpy as np
def solution_iterative_mat(arr, k):
  # mem = [ [0] * k for _ in range(len(arr)+2) ]
  mem = np.zeros((len(arr)+2, k))
  for i in reversed(range(0, len(arr))):
    for c in range(0, k):
      if c+1 == k:
        a = mem[i+2][0]
      else:
        a = mem[i+1][c+1]
      mem[i][c] = max(arr[i] + a, mem[i+1][0])
  return mem
a = solution_iterative_mat([1,2,3,6,1,10], 3) #k=2, solution: 16
```

```
mem = {}
def lose_min_profit(arr, pos, k):
  if pos not in mem:
    res = 0
    if pos < k:
      mem[pos] = 0
    else:
      li = []
      for j in range(pos - k - 1, pos):
        li.append(arr[j] + lose_min_profit(arr, j - 1, k)) 
      res = min(li)
      mem[pos] = res
    # print(f"pos:{pos}, result:{res}")
  # print(f"called pos:{pos}")
  return mem[pos]
arr = [1,2]
sum(arr) - lose_min_profit(arr, 2, 2)
for k,v in mem.items():
  print(k,":",v)
```

```
class MinQueue:

  def __init__(self, maxlen):
    import sys
    self.sysmax = sys.maxsize
    self.s1 = []
    self.s2 = []
    self.maxlen = maxlen
  
  def push(self, i):
    if len(self.s2) + len(self.s1) == self.maxlen:
      self.pop()
    self.s1.append((i, min(i, self.s1[-1][1] if self.s1 else self.sysmax)))

  def get_min(self):
    if not self.s2:
      self.transfer_s1_to_s2()
    return min(self.s1[-1][1] if self.s1 else self.sysmax, self.s2[-1][1] if self.s2 else self.sysmax)
  
  def pop(self):
    if self.s2:
      self.s2.pop()
      if not self.s2:
        self.transfer_s1_to_s2()
    else:
      self.transfer_s1_to_s2()
      self.s2.pop()

  def transfer_s1_to_s2(self):
    while(self.s1):
      element = self.s1.pop()[0]
      min_marker = self.s2[-1][1] if self.s2 else self.sysmax
      self.s2.append((element, min(element, min_marker)))

  def __len__(self):
    return len(self.s1) + len(self.s2)
```

```
def lose_min_profit_using_custom_ds(arr, k):
  mq = MinQueue(k+1)
  for i in arr:
    if len(mq) == k + 1:
      mq.push(mq.get_min() + i)
    else:
      mq.push(i)
    print(mq.s1, " & ", mq.s2, " >>>>> ", mq.get_min())
  return mq.get_min()
arr = [16, 28, 37, 8, 47, 30, 36, 6, 0, 37]
# arr = [1,2,3,1,6,10]
sum(arr) - lose_min_profit_using_custom_ds(arr, 2)
```
