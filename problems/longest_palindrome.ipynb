```
#substring print
def longestPalindrome(s):
    mem = {}
    len_s = len(s) 
    size = 0
    ans_h = 0
    t = 0
    count = 0
    for i in range(0, len_s):
      h = 0
      while h + i < len_s:
        count += 1
        if i == 0:
          mem[h, h+i] = True
        elif i == 1:
          mem[h, h+i] = s[h] == s[h+i]
        else:
          mem[h, h+i] = mem[h+1, h+i-1] and s[h] == s[h+i]
        if mem[h, h+i] and i > size:
          size = i
          ans_h = h
        h = h + 1
    return s[ans_h: ans_h+size+1]
s = "amamamamamaaaaavvvvvv"
longestPalindrome(s)
```
